```python
from transformers import pipeline
from google.colab import userdata
from huggingface_hub import HfApi, login
# Login (you'll be prompted for your token)
hf_token = userdata.get('HF_TOKEN')
print(hf_token)
login(token=hf_token)
# Initialize the pipeline once
pipe = pipeline("text-generation", model="KevinXie0131/my_lora_finetuning_huanhuan_0a", token=hf_token,
    trust_remote_code=True  )

print("=" * 50)
print("甄嬛模拟器已启动 - 输入 'quit' 或 'exit' 退出程序")
print("=" * 50)

while True:
    try:
        # Get user input
        user_input = input("\n请输入你的问题: ")

        # Check for exit condition
        if user_input.lower() in ['quit', 'exit', '退出']:
            print("再见！")
            break

        # Skip empty input
        if not user_input.strip():
            print("请输入有效内容")
            continue

        # Prepare messages
        messages = [
            {"role": "system", "content": "假设你是皇帝身边的女人--甄嬛。"},
            {"role": "user", "content": user_input},
        ]

        # Generate response
        print("\n甄嬛: ", end="")
        response = pipe(messages)

        # Extract and print the generated text
        if response and len(response) > 0:
            generated_text = response[0]['generated_text']
            # Extract only the assistant's response (the last message)
            if isinstance(generated_text, list):
                for msg in generated_text:
                    if msg['role'] == 'assistant':
                        print(msg['content'])
                        break
                else:
                    print(generated_text[-1]['content'] if generated_text else "无法生成回复")
            else:
                print(generated_text)
        else:
            print("无法生成回复")

    except KeyboardInterrupt:
        print("\n\n程序被用户中断")
        break
    except Exception as e:
        print(f"\n发生错误: {e}")
        print("请重试...")
```
