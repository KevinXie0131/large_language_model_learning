```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("KevinXie0131/my_lora_finetuning1")
model = AutoModelForCausalLM.from_pretrained("KevinXie0131/my_lora_finetuning1", device_map="auto")

messages = [
    {"role": "user", "content": "Who was the winner of the 2024 Olympic men’s football tournament?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```
